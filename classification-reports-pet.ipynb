```python
def class_report_file_pet(y_true_csv,pet_dir,outfile, pattern=0):
    """
    y_true: csv file with test example, gold standard, two columns text and label without header
    jsondir: directory with all pet iterations
    pattern: int, pattern to evaluate over
    """
    files = glob.glob(f'{pet_dir}/p{pattern}-i*/predictions.jsonl')
    y_true = pd.read_csv(y_true_csv,names=['text','label']).label.to_list()
    f_scores = []
    reports = ""
    for f in files:
        y_pred = []
        with open(f,'r') as json_file:
            json_list = list(json_file)
            for d in json_list:
                result = json.loads(d)
                y_pred.append(result['label'])
        f_scores.append(f1_score(y_true,y_pred,average='macro'))
        reports+=report(y_true,y_pred)+'\n'
        with open(outfile,'w') as file:
            file.write(reports)
```

```python
class_report_file_pet(y_true_csv='Train-splits/NuclearEnergy_test.csv',pet_dir='pet_full_3pattern/',outfile='pet_reports_roberta-pattern3.txt',pattern=2)
```
